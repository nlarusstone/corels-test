```python
import pandas as pd
import re
```

```python
student = pd.read_csv("../data/student/student_cleaned.csv")
```

```python
pd.set_option('display.max_columns', None)
student.describe(include='all')
```

```python
# Remove students who didn't take SSAT
student = student.ix[student['SSATTotal'].dropna().index]
```

```python
student['Ethnicity'] = student['Ethnicity'].astype('str')
student['Ethnicity'] = student['Ethnicity'].apply(lambda x: 'Black' if re.search('African', x) else x)
student['Ethnicity'] = student['Ethnicity'].apply(lambda x: 'White' if re.search('White|Caucasian', x) else x)
student['Ethnicity'] = student['Ethnicity'].apply(lambda x: 'Asian' if re.search('Asian|Chin|India|Korea|Japan', x) else x)
student['Ethnicity'] = student['Ethnicity'].apply(lambda x: 'Hispanic' if re.search('LatiN|Hispanic|Brazil', x) else x)
student['Ethnicity'] = student['Ethnicity'].apply(lambda x: 'Multiracial' if re.search('Multiracial|Biracial', x) else x)
student['Ethnicity'] = student['Ethnicity'].apply(lambda x: 'Native American' if re.search('Native', x) else x)
student['Ethnicity'] = student['Ethnicity'].apply(lambda x: 'Middle Eastern' if re.search('Middle Eastern', x) else x)
student['Ethnicity'] = student['Ethnicity'].apply(lambda x: 'International' if re.search('International', x) else x)
student['Ethnicity'].unique()
```

```python
student.to_csv("../data/student/student_cleaned.csv")
```
